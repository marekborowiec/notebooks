```python
#! /usr/bin/env python3

# This is a script to convert PartitionFinder v1.x
# scheme block to MrBayes block

fn = 'bs'

# the file contains only scheme lines, e.g.:
# 1      | GTR+I+G    | AA_pos1                  | 1-619\3                 | .\analysis\phylofiles\9e7b447f4c25a86088431b3ef9344d4b.phy
# 2      | GTR+I+G    | AA_pos2, UB_pos2         | 2-620\3, 3151-3778\3    | .\analysis\phylofiles\9ca859df0b1c71238ff37a3599947e86.phy


f = open(fn, 'r')

f_lns = f.readlines()

print('begin mrbayes;')
print('\tset autoclose=yes nowarn=yes;')

tpl_list = []
for line_no, line in enumerate(f_lns):
    line = line.split('|')
    new_line = []
    for index, element in enumerate(line):
        if index == 1 or index == 3:
            element = element.strip(' ')
            new_line.append(element.translate({ord(','): ''}))
    tpl = (line_no, new_line[0], new_line[1])
    tpl_list.append(tpl)
    #print('\t{};'.format(' : '.join(new_line)))
#print(tpl_list)
part_lst = []
for tpl in tpl_list:
    print('\tcharset p{} = {};'.format((tpl[0] + 1), tpl[2]))
    p_name = 'p{}'.format(tpl[0] + 1)
    part_lst.append(p_name)
print('\tpartition currentPartition = {}:{};'.format(len(tpl_list), ','.join(part_lst)))
print('\tset partition = currentPartition;')

uniq = list(set([tpl[1] for tpl in tpl_list]))
for model in uniq:
    appl = [str(tpl[0] + 1) for tpl in tpl_list if tpl[1] == model]
    if model == 'GTR+I+G':
        print('\tlset applyto=({}) nst=6 rates=invgamma;'.format(','.join(appl)))
    elif model == 'SYM+I+G':
        print('\tlset applyto=({}) nst=6 rates=invgamma;'.format(','.join(appl)))
        print('\tprset applyto=({}) statefreqpr=fixed(equal);'.format(','.join(appl)))
print('\tprset applyto=(all) ratepr=variable brlenspr=unconstrained:gammadir(1,10,1,1);')
print('\tmcmc ngen=25000000 relburnin=yes burninfrac=0.20 printfreq=2500 samplefreq=2500 nruns=2 nchains=4 savebrlens=yes;')
print('\tsump;')
print('\tsumt;')
print('end;')
```
