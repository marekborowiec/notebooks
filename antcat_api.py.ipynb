```python
#! /usr/bin/env python
# antcat_api.py - antcat.org API inteface sandbox

import requests
import json

# this gest first 100 reference records as a single json string
r = requests.get("http://antcat.org/v1/references")

# this gets all journals
jr = requests.get("http://antcat.org/v1/journals")

refs = r.text
journals = jr.text

# parse json refs
parsed_refs = json.loads(refs)
# parse json journals
parsed_journals = json.loads(journals)

# get a list of dicts where reference type is one of the elements
new_journals = [{'type' : list(rec.keys())[0]} for rec in parsed_journals]

new_journals = []
for rec in parsed_journals:
    rec_dict = list(rec.values())[0]
    rec_type = list(rec.keys())[0]
    rec_dict['type'] = rec_type
    new_journals.append(rec_dict)
    
    
def get_jour_name(id):
    for journal in new_journals:
        if journal['id'] == id:
            name = journal['name']
            return name
        
def get_bib_dict(reference):
    ref_type, ref_dict in reference.items()
    bib_dict = {'ref_type': ref_type, 
               'author' : ref_dict['author_names_string_cache'],
               'title' : ref_dict['title'],
               'year' : ref_dict['year'],
               'journal' : get_jour_name(ref_dict['journal_id']),
               'volume' : ref_dict['series_volume_issue'].split('(')[0]}
    return bib_dict
    

# the parsed contents are stored as a list of nested dicts
#print(parsed_refs)

for ref in parsed_refs:
    # correct
    print(ref)
    # why does this pull out only last reference?
    print(get_bib_dict(ref))
    # get tuple that has reference type as zeroth item
    # and dict with all reference fields in first item
    #for ref_type, ref_dict in ref.items():
        #print(ref_type)
        #print(sorted(ref_dict.keys()))
        #print(ref_dict)
        #print(ref_dict['author_names_string_cache'])
        #print(ref_dict['year'])
        #print(ref_dict['title'])
        #print(get_jour_name(ref_dict['journal_id']))
        #print(ref_dict['series_volume_issue'])
        #print(ref_dict['pagination'])
```
