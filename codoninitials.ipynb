```python
import csv, re, requests, string

# download MIT Pantheon data
download_url = 'http://pantheon.media.mit.edu/pantheon.tsv'

res = requests.get(download_url)

res_str = res.text
# turn data into list of rows
res_table = res_str.splitlines()

# these are our codons
codons = 'ACGTacgt'

# turn data rows into a lists,
# one element per column
row_list = []
for row in res_table:
    row = row.split('\t')
    row_list.append(row)

# print header
#print(row_list[0])

# skip header
for row in row_list[1:]:
    inits = []
    # we know that name is the second element
    name = row[1]
    # split into a list by a non-word character
    name = re.findall(r'[\w]+', name)
    # keep track of how many components in name
    counter = 0
    # check if codons match
    for element in name:
        for char in codons:
            if element.startswith(char):
                inits.append(element[0])
                counter += 1
    # print if all elements match codon letter
    if counter == len(name) and len(name) >= 2:
        codon_init = ''.join(inits).translate({ord(i) : None for i in string.ascii_lowercase})
        codon_name = ' '.join(name)
        print('{} : {}'.format(codon_name, codon_init))
```
